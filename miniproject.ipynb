# moving average of various stocks

```python
google1 = google['Close'].to_frame()
google1['mvngavg']=google1['Close'].rolling(7).mean()
google1.dropna(inplace=True)
microsoft1 = microsoft['Close'].to_frame()
microsoft1['mvngavg']=google1['Close'].rolling(7).mean()
microsoft1.dropna(inplace=True)
apple1 = apple['Close'].to_frame()
apple1['mvngavg']=google1['Close'].rolling(7).mean()
apple1.dropna(inplace=True)
amazon1 =amazon['Close'].to_frame()
amazon1['mvngavg']=google1['Close'].rolling(7).mean()
amazon1.dropna(inplace=True)
f,ax=plt.subplots(2,2,figsize=(100,30))
scs.lineplot(x='Date',y='Close',data=google1,ax=ax[0,0],label='close').set(ylabel=None,title='Google')
scs.lineplot(x='Date',y='mvngavg',data=google1,ax=ax[0,0],label='mvngavg').set(ylabel=None)
scs.lineplot(x='Date',y='Close',data=microsoft1,ax=ax[0,1],label='close').set(ylabel=None,title='microsoft')
scs.lineplot(x='Date',y='mvngavg',data=microsoft1,ax=ax[0,1],label='mvngavg').set(ylabel=None)
scs.lineplot(x='Date',y='Close',data=apple1,ax=ax[1,0],label='close').set(ylabel=None,title='apple')
scs.lineplot(x='Date',y='mvngavg',data=apple1,ax=ax[1,0],label='mvngavg').set(ylabel=None)
scs.lineplot(x='Date',y='Close',data=amazon1,ax=ax[1,1],label='close').set(ylabel=None,title='amazon')
scs.lineplot(x='Date',y='mvngavg',data=amazon1,ax=ax[1,1],label='mvngavg').set(ylabel=None)
```

# daily return average of a stock

```python
google1 = google['Close'].to_frame()
google1['preclose']=google1['Close'].shift(1)
google1.dropna(inplace=True)
google1['return']= (google1['Close'] / google1['preclose']) - 1
microsoft1 = microsoft['Close'].to_frame()
microsoft2=microsoft1
microsoft1['preclose']=microsoft1['Close'].shift(1)
microsoft1.dropna(inplace=True)
microsoft1['return']= (microsoft1['Close'] / microsoft1['preclose'])
apple1 = apple['Close'].to_frame()
apple2=apple1
apple1['preclose']=apple1['Close'].shift(1)
apple1.dropna(inplace=True)
apple1['return']= (apple1['Close'] / apple1['preclose'])
amazon1 =amazon['Close'].to_frame()
amazon1['preclose']=amazon1['Close'].shift(1)
amazon1.dropna(inplace=True)
amazon1['return']= (amazon1['Close'] / amazon1['preclose'])
f,ax=plt.subplots(2,2,figsize=(100,60))
scs.lineplot(x='Date',y='return',data=google1,ax=ax[0,0],label='google1').set(ylabel='Date',title='Google')
scs.lineplot(x='Date',y='return',data=microsoft1,ax=ax[0,1],label='microsoft1').set(ylabel='Date',title='microsoft')
scs.lineplot(x='Date',y='return',data=apple1,ax=ax[1,0],label='apple1').set(ylabel='Date',title='apple')
scs.lineplot(x='Date',y='return',data=amazon1,ax=ax[1,1],label='amazon1').set(ylabel='Date',title='amazon')
```

# Adding new column â€˜Trendâ€™ whose values are based on the 'Daily Return'

```python
google['trend']=google1['return']
microsoft['trend']=microsoft1['return']
apple['trend']=apple1['return']
amazon['trend']=amazon1['return']
```

```python
trend=pd.DataFrame({'company':['google','microsoft','apple','amazon'],'trend':[google['trend'].mean(),microsoft['trend'].mean(),apple['trend'].mean(),amazon['trend'].mean()]})
```

# Visualising trend frequency through a Pie Chart

```python

plt.pie(trend['trend'],labels=trend['company'],autopct='%1.5f%%', textprops={'fontsize': 14})
plt.labelsize=(5)
```

# correlation between the daily returns of different stocks

```python
Returns=pd.DataFrame({'GOGL':google1['return'],'MSFT':microsoft1['return'],'AAPL':apple1['return'],'AMZN':amazon1['return']})
Returns.dropna(inplace=True)
f,ax=plt.subplots(figsize=(10,10))
scs.heatmap(Returns.corr(),annot=True,linewidth=0.5,linecolor='black',fmt='.1f',ax=ax)
```
